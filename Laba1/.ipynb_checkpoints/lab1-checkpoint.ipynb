```python
import pandas
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.utils import to_categorical 
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe = pandas.read_csv("iris.csv", header=None) 
dataset = dataframe.values 
X = dataset[:,0:4].astype(float) 
Y = dataset[:,4]
```

```python
encoder = LabelEncoder() 
encoder.fit(Y) 
encoded_Y = encoder.transform(Y) 
dummy_y = to_categorical(encoded_Y)
```

```python
model_ex = Sequential() 
model_ex.add(Dense(4, activation='relu')) 
model_ex.add(Dense(3, activation='softmax'))
```

```python
model_ex.compile(optimizer='adam',loss='categorical_crossentropy',  metrics=['accuracy'])
```

```python
history_ex = model_ex.fit(X, dummy_y, epochs=75, batch_size=10,  validation_split=0.1)
```

```python
mymodel = Sequential()
mymodel.add(Dense(3, activation='elu'))
mymodel.add(Dense(4, activation='relu')) 
mymodel.add(Dense(3, activation='softmax'))
```

```python
mymodel.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
mymodel_his = mymodel.fit(X, dummy_y, epochs=75, batch_size=10,  validation_split=0.1)
```

```python
copymymodel = Sequential()
copymymodel.add(Dense(3, activation='elu'))
copymymodel.add(Dense(4, activation='relu')) 
copymymodel.add(Dense(3, activation='softmax'))
```

```python
copymymodel.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
copymymodel_his = copymymodel.fit(X, dummy_y, epochs=75, batch_size=15,  validation_split=0.1)
```

```python
plt.plot(history_ex.history['accuracy'])
plt.plot(mymodel_his.history['accuracy'])
plt.plot(copymymodel_his.history['accuracy'])
plt.legend(['  ','    elu','     '])
```

```python
plt.plot(history_ex.history['loss'])
plt.plot(mymodel_his.history['loss'])
plt.plot(copymymodel_his.history['loss'])
plt.legend(['  ','    elu','     '])
```
